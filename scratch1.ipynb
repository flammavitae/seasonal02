```python
import yfinance as yf
import pandas as pd
import numpy as np

# Define the tickers for S&P 500, Nasdaq 100, Gold, and US Bonds
tickers = ['^GSPC', '^NDX', 'GC=F', 'TLT']

# Fetch historical data for as long as possible
data = yf.download(tickers, period='max')['Close']

# Resample data to quarterly frequency and calculate returns
quarterly_returns = data.resample('QE').ffill().pct_change().dropna()

# Define a function to calculate the best portfolio for each quarter
def calculate_best_portfolio(returns):
    num_assets = len(returns.columns)
    num_portfolios = 10000
    results = np.zeros((num_portfolios, num_assets + 2))

    for i in range(num_portfolios):
        weights = np.random.random(num_assets)
        weights /= np.sum(weights)
        portfolio_return = np.sum(weights * returns.mean()) * 4
        portfolio_stddev = np.sqrt(np.dot(weights.T, np.dot(returns.cov() * 4, weights)))
        results[i, :num_assets] = weights
        results[i, num_assets] = portfolio_return
        results[i, num_assets + 1] = portfolio_stddev

    results_df = pd.DataFrame(results, columns=[*returns.columns, 'Return', 'Volatility'])
    best_portfolio = results_df.iloc[results_df['Return'].idxmax()]
    return best_portfolio

# Calculate the best portfolio for each quarter
best_portfolios = quarterly_returns.groupby(quarterly_returns.index).apply(calculate_best_portfolio)

# Display the best portfolios
print(best_portfolios)
```
